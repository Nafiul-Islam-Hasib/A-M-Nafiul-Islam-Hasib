
<h1 style = "color : blue ; " > Manullay computed result for slope(M) and intercept(C)</h1>


```python
# mean_x = mean of numpy array x 
# mean_y = mean of numpy array y 
mean_x = np.mean(x)
mean_y = np.mean(y)

# x_ = xi-x^
# y_ = yi-y^
x_ = x - mean_x
y_ = y - mean_y 

nemenater = np.sum(x_*y_)
denemenater = np. sum(x_**2)

M = nemenater/denemenater
print(M)

C = mean_y - M*mean_x
print(C)
```


#### Predicted values for FINAL BEST FIT LINE 


```python
Y = M*x + C
Y
```


#### Result Calculation for weight = 6.50


```python
Y_value = M*6.50 + C
Y_value
```


#### MEAN SQUARE ERROR ( MSE )


```python
MSE = np.sum((y-Y)**2)/len(x)
MSE
```

#### MEAN ABSOLUTE ERROR (MAE)


```python
MEAN = np.sum(abs(y-Y))/len(x)
MEAN
```


<h1 style = "color : blue ;">Linear Regression with Sk-learn</h1>


```python
from sklearn.linear_model import LinearRegression 
reg = LinearRegression()
```

```python
reg.fit(df[['Weight']],df['Price'])
```

```python
M = reg.coef_
M
```

```python
C = reg.intercept_
C
```


<h1 style="color:green;font-size:20px;">predicted Target for 6.50</h1>
                 

```python
reg.predict([[6.50]])
```

<h1 style = "color: green; font-size: 20px; "> Predicted value for Target y </h1>

```python
Target_y  = f"{float(M):.2f} * {x} + {float(C):.2f}"
Target_y
```

<h1 style = "color: green; font-size: 20px; "> plot for best fit line</h1>

```python
reg.score(df[['Weight']],df['Price'])
```
